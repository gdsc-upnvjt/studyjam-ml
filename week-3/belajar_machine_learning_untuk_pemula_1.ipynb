# **Decision Tree Implementation**

## Data Gathering

```python
import pandas as pd

iris = pd.read_excel("https://docs.google.com/spreadsheets/d/1RAVy17NP-Fr08pIXBfEUWQ2ijYg5C2b3NTcO7c7ijNY/export?gid=0&format=xlsx")
iris.head()
```

```python
# Melihat informasi dataset
iris.info()

# melihat informasi dataset pada 5 baris pertama
iris.head()
```

## Data Prepration

```python
iris.drop('Id',axis=1,inplace=True)
```

## Data Preprocessing

```python
# memisahkan atribut dan label
X = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm' ]]
y = iris['Species']

# Membagi dataset menjadi data latih & data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

## Modelling

```python
from sklearn.tree import DecisionTreeClassifier

# membuat model Decision Tree
tree_model = DecisionTreeClassifier()

# Melatih model dengan menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)
```

## Model Evaluation

```python
# Evaluasi Model
from sklearn.metrics import accuracy_score

y_pred = tree_model.predict(X_test)

acc_secore = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_secore)
```

## Model Inference

```python
# prediksi model dengan tree_model.predict([[SepalLength, SepalWidth, PetalLength, PetalWidth]])
result = tree_model.predict([[6.2, 3.4, 5.4, 2.3]])[0]
print("Jenis Iris untuk data [6.2, 3.4, 5.4, 2.3] : ", result)
```

## **How Linear Regression Works**

```python
import numpy as np
from IPython.display import clear_output
class linear_regression:
    def __init__(self, learning_rate = 0.01, iteration = 40):
        self.learning_rate = learning_rate
        self.iteration = iteration
        self.bias = 0

    def fit(self, features, labels):
        self.m, self.n = features.shape
        self.weights = np.zeros(self.n)

        for i in range(self.iteration):
          y_predict = self.predict(features)
          cost = (1/(2*self.m)) * np.sum((y_predict - labels)**2)

          derevative_weight = (1/self.m) * np.dot(features.T, y_predict-labels )
          derevative_bias =  (1/self.m) * np.sum(y_predict-labels)

          self.weights = self.weights - (self.learning_rate * derevative_weight)
          self.bias = self.bias - (self.learning_rate * derevative_bias)

          clear_output(wait=True)
          plt.title(f'Iteration {i+1}')
          a = np.array(x)
          b = self.linear_regression_formula(x, self.weights[0], self.bias)

          plt.scatter(x, y, c='red')
          plt.plot(a, b)

          for i in range(0, len(x)):
            plt.plot([x[i],a[i]], [y[i],b[i]], 'r--')

          plt.show()

        return self

    def linear_regression_formula(self, x, m, c):
        return m*x+c
    def predict(self, X):
        return np.dot(X, self.weights ) + self.bias
```

```python
x = np.array([1,1,2,2,3,4,4,5,5,5])
y = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000,85000, 90000])
x_reshaped = x.reshape(-1,1)
linreg = linear_regression()
linreg.fit(x_reshaped, y)
```

### Logistic Regression Works Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
class LogisticRegression:
  def __init__(self, learning_rate=0.01, iterations=50):
    self.learning_rate = learning_rate
    self.iterations = iterations
    self.bias = 0

  def sigmoid(self, z):
    return 1 / (1 + np.exp(-z))

  def fit(self, features, labels):
    self.m, self.n = features.shape
    self.weights = np.zeros(self.n)

    for i in range(self.iterations):
      z = np.dot(features, self.weights) + self.bias
      y_predict = self.sigmoid(z)

      cost = (-1/self.m) * np.sum(labels * np.log(y_predict) + (1 - labels) * np.log(1 - y_predict))

      derivative_weight = (1/self.m) * np.dot(features.T, (y_predict - labels))
      derivative_bias = (1/self.m) * np.sum(y_predict - labels)

      self.weights -= self.learning_rate * derivative_weight
      self.bias -= self.learning_rate * derivative_bias
      clear_output(wait=True)
      m = self.weights[0]
      c = self.bias
      plt.scatter(x, y, c='red')
      a = np.array(x)
      b = self.my_formula(a, m, c)
      plt.title(f'Iteration {i+1}')
      plt.plot(a, b)
      for i in range(0, len(x)):
        plt.plot([x[i],a[i]], [y[i],b[i]], 'r--')
      plt.show()

    return self

  def my_formula(self, x, m, c):
    z = m * x + c
    return 1 / (1 + np.exp( - z))

  def predict(self, X):
    z = np.dot(X, self.weights) + self.bias
    return self.sigmoid(z)
```

# **K-Means Implementation**

## Data Gathering

```python
customer = pd.read_excel("https://docs.google.com/spreadsheets/d/1Cl6aNAMWslqmNJ3i1nKPPHVW3fDJ6F4UdNoC0b2tWrQ/export?gid=0&format=xlsx")
```

## Data Preprocessing

```python
# ubah nama kolom
customer = customer.rename(columns={'Gender': 'gender', 'Age': 'age',
                        'Annual Income (k$)': 'annual_income',
                        'Spending Score (1-100)': 'spending_score'})

# tampilkan data yang sudah di preprocess
customer.head(3)
```

## Implement Elbow Method to Found Optimal K

```python
from sklearn.cluster import KMeans

# menghilangkan kolom customer id dan gender
X = customer.drop(['CustomerID', 'gender', 'annual_income'], axis=1)

#membuat list yang berisi inertia
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

## Modelling

```python
km5 = KMeans(n_clusters=3).fit(X)

# menambahkan kolom label pada dataset
X['Labels'] = km5.labels_

# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(x=X['age'], y=X['spending_score'], hue=X['Labels'],
                palette=sns.color_palette('hls', 3))
plt.title('KMeans dengan 3 Cluster')
plt.show()
```
